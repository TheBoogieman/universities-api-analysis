```python
import duckdb

# Connect to the database
con = duckdb.connect('../data/universities.duckdb')

# View Task 3 results
con.execute("""
    SELECT * FROM uk_france_china_counts
""").df()

# Explore all universities
con.execute("""
    SELECT country, COUNT(*) as count
    FROM all_universities
    GROUP BY country
    ORDER BY count DESC
    LIMIT 10
""").df()
```

```python
# Task 5a: View enrollment percentages for 2010 and 2013
import duckdb

# Connect to the database
con = duckdb.connect('../data/universities.duckdb')

# View Task 3 results
con.execute("""
    SELECT * FROM uk_france_china_counts
""").df()

# Explore all universities
con.execute("""
SELECT 
    year,
    FORMAT('{:,}',ROUND(enrollment_rate, 6)) AS enrollment_rate,
    FORMAT('{:,}',CAST(total_enrolled AS BIGINT)) AS total_enrolled,
    FORMAT('{:,}',population) as population,
    total_enrolled_calculated,
    country,
    iso_code
FROM enrollment_percentages;
""").df()
```

```python
# Task 5b: View global enrollment percentages for 2010
import duckdb

# Connect to the database
con = duckdb.connect('../data/universities.duckdb')

# View Task 3 results
con.execute("""
    SELECT * FROM uk_france_china_counts
""").df()

# Explore all universities
con.execute("""
    SELECT *
    FROM highest_enrollment_global_2010 
    order by total_enrolled desc limit 20
""").df()
```
